```python
import geopandas as gpd
import pandas as pd
file = gpd.read_file('/Users/sohampradhan/Downloads/denver/existing_landuse_2018.shp')
file2 = gpd.read_file('/Users/sohampradhan/Downloads/tl_2021_08_tract/tl_2021_08_tract.shp')
file3 = file.sjoin(file2, how="inner", predicate='intersects')
file4 = file3
file4 = file4.reset_index()
for i in range(186156):
    if('Other/Unknown' in str(file4['CPD_LANDUS'][i])):
        file4 = file4.drop(i)
file4 = file4.loc[:, file4.columns.intersection(['GEOID','CPD_LANDUS'])]
#print(file4)
file4 = file4.groupby(file4.columns.tolist(),as_index=False).size()
pd.set_option('display.max_columns', None)
file4['Count_Column'] = file4.groupby(['GEOID'])['size'].transform('sum')
#print(file4.loc[file4['NAME'] == '88.02'])
file4["size"] = file4["size"]/file4["Count_Column"]
#print(file4)
del file4['Count_Column']
#print(file4)
pivoted1 = file4.pivot(index='GEOID', columns='CPD_LANDUS', values='size')\
            .reset_index()
pivoted1.columns.name=None
pivoted1 = pivoted1.fillna(0)
pivoted1 = pivoted1.rename(columns={'GEOID':'DenverCT','Single-unit Residential': 'Single_unit_Residential_land_ratio', 'Multi-unit Residential': 'Multi_unit_Residential_land_ratio', 'Vacant': 'Vacant_land_ratio', 'Industrial': 'Industrial_land_ratio', 'Two-unit Residential': 'Two_unit_Residential_land_ratio', 'Commercial/Retail': 'Commercial', 'Office': 'Office', 'Parking': 'Parking_land_ratio', 'Public/Quasi-public': 'Public/Quasi-public_land_ratio','Park/Open Space':'Park/Open_Space_land_ratio','Mixed-use':'Mixed_use','Trans/Comm/Utilities':'Trans/Comm/Utilities_land_ratio','ROW/Road':'ROW/Road_land_ratio','Agriculture':'Agriculture','Entertainment/Cultural':'Entertainment/Cultural_land_ratio'})
pivoted1['Residential'] = pivoted1['Single_unit_Residential_land_ratio']+pivoted1['Multi_unit_Residential_land_ratio'] + pivoted1['Two_unit_Residential_land_ratio']
pivoted1['Industrial'] = pivoted1['Industrial_land_ratio']+pivoted1['Trans/Comm/Utilities_land_ratio']
pivoted1['Parks/Open Space'] = pivoted1['Vacant_land_ratio']+pivoted1['ROW/Road_land_ratio']+pivoted1['Parking_land_ratio'] + pivoted1['Park/Open_Space_land_ratio']
pivoted1['Institutional'] = pivoted1['Public/Quasi-public_land_ratio']+pivoted1['Entertainment/Cultural_land_ratio']
del pivoted1['Single_unit_Residential_land_ratio']
del pivoted1['Multi_unit_Residential_land_ratio']
del pivoted1['Two_unit_Residential_land_ratio']
del pivoted1['Industrial_land_ratio']
del pivoted1['Trans/Comm/Utilities_land_ratio']
del pivoted1['Vacant_land_ratio']
del pivoted1['ROW/Road_land_ratio']
del pivoted1['Parking_land_ratio']
del pivoted1['Public/Quasi-public_land_ratio']
del pivoted1['Entertainment/Cultural_land_ratio']
del pivoted1['Park/Open_Space_land_ratio']
pivoted1.head()
pivoted1.to_csv(r"/Users/sohampradhan/Downloads/denver+columbus/denver_ct_ratios2015.csv", index = False)
```
